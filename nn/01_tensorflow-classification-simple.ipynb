```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Input, Dense
import numpy as np
```

```python
X_train = np.array([[0.1, 0.2], [0.2, 0.1], [0.3, 0.2], [0.4, 0.35], [0.5, 1.3], [1.3, 1.4], [1.2, 1.8], [2.0, 1.2]])
y_train = np.array([0, 0, 0, 0, 1, 1, 1, 1])
```

```python
model = Sequential([
    Input(shape=(2,)),                      # explicit input
    Dense(3, activation='relu'),            # hidden
    Dense(1, activation='linear')           # raw logits
], name='binary_classifier')
```

```python
model.summary()
```

```python
tf.random.set_seed(42)  # for reproducibility
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss=tf.keras.losses.BinaryCrossentropy(from_logits=False), metrics=['accuracy']) # from_logits -> Numerically more stable
model.fit(X_train, y_train, epochs=300)
```

```python
X = np.array([[0.15, 0.25]])
logit = model.predict(X)              # raw logit
prob  = tf.nn.sigmoid(logit).numpy()  # convert to probability
print(f"Predicted probability for {X} is {prob[0][0]:.4f} to be of class 1")
print(f"Predicted class: {1 if prob[0][0] > 0.5 else 0}")
```
